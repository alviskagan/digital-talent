## 1. Deskripsi Data

### Tanaman Padi per Provinsi
Dataset ini berisi mengenai luas lahan, produksi, dan produktivitas dari tanaman padi per provinsi

Penjelasan mengenai Variabel pada Dataset ini:

1. tahun: Tahun
2. kode_provinsi: Kode Provinsi
3. nama_provinsi: Nama Provinsi
4. luas_panen: Luas Panen dari Tanaman Pangan (dalam Hektar)
5. produktivitas: Produktivitas Tanaman Pangan (dalam Kuintal/Hektar)
6. produksi: Total Produksi Tanaman Pangan (dalam Ton)
7. latitude: koordinat lintang lokasi provinsi
8. longitude: koordinat bujur lokasi provinsi

Luas panen adalah luasan tanaman yang dipungut hasilnya setelah tanaman tersebut cukup umur. Bentuk Produksi Jagung meliputi jagung pipilan kering.

```python
import pandas as pd
data_tanaman = pd.read_csv('data-tanaman-pangan-padi-perprovinsi-1993-2013.csv', delimiter=',')
```

## 2. Statistik Tentang Data

```python
data_tanaman.median()
data_tanaman.max()
```

```python
data_tanaman.min()
```

```python
# Rata-Rata Produksi Padi Selama 20 Tahun (Ton)
pdk = data_tanaman.produksi.mean()
```

```python
# Rata-Rata Produktivitas Padi Selama 20 Tahun (Kuintal/Hektar)
pdv = data_tanaman.produktivitas.mean()
```

```python
# Rata-Rata Luas Daerah Panen Selama 20 Tahun (Hektar)
lp = data_tanaman.luas_panen.mean()
```

```python
print("Rata-Rata produksi padi dari tahun 1993 - 2013 sejumlah: ", pdk, "Ton" )
print("Rata-Rata produktivitas padi dari tahun 1993 - 2013 sejumlah: ", pdv, "Kuintal/ Hektar" )
print("Rata-Rata luas daerah panen padi dari tahun 1993 - 2013 sejumlah: ", lp, "Hektar" )
```

```python
data_tanaman.groupby(('tahun')).mean()
```

# 3. Visualisasi Data

```python
ax = data_tanaman.groupby('tahun').mean()['produksi']
ax = ax.plot(kind='line', title ="Rata-Rata Berdasarkan Tahun", figsize=(20, 10), legend=True, fontsize=12)
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Produksi", fontsize=12)
```

```python
ax = data_tanaman.groupby('nama_provinsi').mean()['produktivitas']
ax = ax.plot(kind='bar', title ="Rata-Rata Produksi Padi Setiap Provinsi Selama 20 Tahun ", figsize=(20, 10), legend=True, fontsize=12)
ax.set_xlabel("Provinsi", fontsize=12)
ax.set_ylabel("Produktivitas", fontsize=12)
```

# 4. Narasi Analisis
