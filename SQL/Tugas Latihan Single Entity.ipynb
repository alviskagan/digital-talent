```python
import os
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

```python
# 1. Prepare a list of offices sorted by country, state, city
list_offices = run("SELECT * FROM offices ORDER BY country, state, city")
list_offices
```

```python
# 2. How many employees are there in the company?
count_emp = run("SELECT COUNT(employeeNumber) as Total_Employee FROM employees")
count_emp
```

```python
# 3. What is the total of payments received?
total_pay = run("SELECT SUM(amount) as Total_Payment FROM payments")
total_pay
```

```python
# 4. List the product lines that contain 'Cars'.
product_cars = run("SELECT * FROM products WHERE productLine LIKE '%Cars'")
product_cars
```

```python
# 5. Report total payments for October 28, 2004.
pay_oct = run("SELECT SUM(amount) as Total_Payment_October FROM payments WHERE paymentDate = '2004-10-28'")
# pay_oct = run("SELECT * FROM payments")
pay_oct
```

```python
# 6. Report those payments greater than $100,000.
gt_100 = run("SELECT * FROM payments WHERE amount > 100000")
gt_100
```

```python
# 7. List the products in each product line.
product_line = run("SELECT * FROM products ORDER BY productLine")
product_line
```

```python
# 8. How many products in each product line?
total_product_perline = run("SELECT productLine, COUNT(productCode) as Total_Product FROM products GROUP BY productLine")
total_product_perline
```

```python
# 9. What is the minimum payment received?
min_pay = run("SELECT * FROM payments WHERE amount HAVING MIN(amount)")
min_pay
```

```python
# 10. List all payments greater than twice the average payment.
gt_twice_avg = run("SELECT * FROM payments WHERE amount > (SELECT AVG(amount)*2 FROM Payments)") 
gt_twice_avg
# run("select * from payments where amount > AVG(amount)*2")
```

```python
# 11. What is the average percentage markup of the MSRP on buyPrice?
run("""
SELECT
    (AVG(MSRP - buyPrice)/buyPrice)*100 as Result
FROM
    productS
""")
```

```python
# 12. How many distinct products does ClassicModels sell?
run("""
SELECT
    DISTINCT(productName)
FROM
    products
""")
```

```python
# 13. Report the name and city of customers who don't have sales representatives?
run("""
SELECT
    customerName, city, salesRepEmployeeNumber
FROM
    customers
WHERE
    salesRepEmployeeNumber IS NULL
""")
```

```python
# 14. What are the names of executives with VP or Manager in their title? 
# Use the CONCAT function to combine the employee's first name and last name 
# into a single field for reporting.
run("""
SELECT 
    CONCAT(firstName, ' ', lastName) as Full_Name, jobTitle
FROM
    employees
WHERE
    jobTitle like "VP%" or jobTitle like "%Manager%"
""")
```

```python
# 15. Which orders have a value greater than $5,000?
run("""
SELECT
    productName, msrp, priceEach, (msrp-priceEach) as Value
FROM
    orders o 
    INNER JOIN
    orderdetails od USING(orderNumber)
    INNER JOIN
    products p USING(productCode)
GROUP BY productName
    HAVING Value > 5 
""")
```
